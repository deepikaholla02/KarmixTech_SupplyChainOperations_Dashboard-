```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv("DataCoSupplyChainDataset.csv",encoding='latin1')
df.head()
```

```python
df = df.dropna(subset=['Customer Lname', 'Customer Zipcode'])
```

```python
df.isna().sum()
```

```python
print("Shape After Removing Nulls:", df.shape)
```

```python
# Check duplicates
print("Duplicate Rows:", df.duplicated().sum())
```

```python
# DROP UNNECESSARY COLUMNS
drop_cols = [
    'Customer Email',
    'Customer Password',
    'Customer Street',
    'Customer Fname',
    'Customer Lname',
    'Product Description',
    'Product Image'
]

df = df.drop(columns=drop_cols)
```

```python
# CONVERT DATE COLUMNS

df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])
```

```python
# FEATURE ENGINEERING

# Delivery Time
df['Delivery Time'] = (df['shipping date (DateOrders)'] -df['order date (DateOrders)']).dt.days

# Profit Margin %
df['Profit Margin %'] = (df['Order Profit Per Order'] / df['Sales']) * 100

# Month & Year
df['Order Month'] = df['order date (DateOrders)'].dt.month
df['Order Year'] = df['order date (DateOrders)'].dt.year
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Order Profit Per Order'].sum()
total_orders = df['Order Id'].nunique()
avg_delivery = df['Delivery Time'].mean()
late_delivery_percent = (df['Late_delivery_risk'].mean()) * 100

print("Total Sales:", round(total_sales, 2))
print("Total Profit:", round(total_profit, 2))
print("Total Orders:", total_orders)
print("Average Delivery Time:", round(avg_delivery, 2))
print("Late Delivery %:", round(late_delivery_percent, 2))
```

```python
# EXPLORATORY DATA ANALYSIS (EDA)

# SALES BY MARKET
market_sales = (df.groupby('Market')['Sales'].sum().sort_values(ascending=False))
plt.figure(figsize=(8,5))
sns.barplot(x=market_sales.index,y=market_sales.values)
plt.title('Sales by Market')
plt.xlabel('Market')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# SALES BY CATEGORY

category_sales = (df.groupby('Category Name')['Sales'].sum().sort_values(ascending=False))
plt.figure(figsize=(12,8))
sns.barplot(x=category_sales.values,y=category_sales.index)
plt.title('Sales by Category')
plt.xlabel('Sales')
plt.ylabel('Category')
plt.show()
```

```python
# TOP 10 PRODUCTS
top_products = (df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10))
plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values,y=top_products.index)
plt.title('Top 10 Products')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.show()
```

```python
# DELIVERY STATUS ANALYSIS

plt.figure(figsize=(8,5))
sns.countplot(x='Delivery Status',data=df)
plt.title('Delivery Status Distribution')
plt.xticks(rotation=20)
plt.show()
```

```python
# LATE DELIVERY ANALYSIS

late_delivery = (df['Late_delivery_risk'].value_counts())
plt.figure(figsize=(8,5))
plt.pie(late_delivery,labels=['No Risk', 'Late Risk'],autopct='%1.1f%%')
plt.title('Late Delivery Risk')
plt.show()
```

```python
# SHIPPING MODE PERFORMANCE

shipping_perf = (df.groupby('Shipping Mode')['Delivery Time'].mean().sort_values())
plt.figure(figsize=(10,5))
sns.barplot(x=shipping_perf.index,y=shipping_perf.values)
plt.title('Average Delivery Time by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Delivery Time')
plt.show()
```

```python
# MONTHLY SALES TREND

monthly_sales = (df.groupby('Order Month')['Sales'].sum())
plt.figure(figsize=(12,6))
monthly_sales.plot(marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# PROFIT BY MARKET
market_profit = (df.groupby('Market')['Order Profit Per Order'].sum().sort_values(ascending=False))
plt.figure(figsize=(12,6))
sns.barplot(x=market_profit.index,y=market_profit.values)
plt.title('Profit by Market')
plt.xlabel('Market')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
# CUSTOMER SEGMENT ANALYSIS

segment_sales = (df.groupby('Customer Segment')['Sales'].sum())
plt.figure(figsize=(8,5))
sns.barplot(x=segment_sales.index,y=segment_sales.values)
plt.title('Sales by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Sales')
plt.show()
```

```python
# ORDER STATUS ANALYSIS

plt.figure(figsize=(10,5))
sns.countplot(x='Order Status',data=df)
plt.title('Order Status')
plt.xticks(rotation=45)
plt.show()
```

```python
# CORRELATION HEATMAP

numeric_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(14,10))
sns.heatmap(numeric_df.corr(),cmap='coolwarm',annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# TOP REGIONS BY SALES

region_sales = (df.groupby('Order Region')['Sales'].sum().sort_values(ascending=False).head(10))
plt.figure(figsize=(12,6))
sns.barplot(x=region_sales.values,y=region_sales.index)
plt.title('Top Regions by Sales')
plt.xlabel('Sales')
plt.ylabel('Region')
plt.show()
```

```python
# PROFIT MARGIN DISTRIBUTION

plt.figure(figsize=(10,5))
sns.histplot(df['Profit Margin %'],bins=30,kde=True)
plt.title('Profit Margin Distribution')
plt.xlabel('Profit Margin %')
plt.show()
```

```python
# SAVE Analyzed DATASET

df.to_csv('supply_chain_dataset.csv',index=False)
print("\nDataset Saved Successfully")
```
